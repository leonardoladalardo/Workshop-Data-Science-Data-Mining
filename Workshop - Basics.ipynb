## Section 2: What is Data Science and Data Mining?

If we go to wikipedia, how will it define **Data Science**?

- "Data science is a multi-disciplinary field that uses **scientific methods, processes, algorithms and systems** to extract **knowledge and insights** from data in various forms, both **structured and unstructured**,and largely **synonymous to data mining and big data.**"

However, there is a common mistake while refering **Data Mining and Big Data as a synonym of Data Science**. As we're going to see, Data Mining and Big Data consists in a branch of Data Science, as they're based on what we call as Data Exploration tools (algorithms to collect data) and forms (size and dimensionality). 

```python
#Branchs of Data Science:
from IPython.display import Image
Image('https://cdn3.datascience.berkeley.edu/content/0be56de3069740669fe2d696fb60220e/DataScienceLifeCycle.jpg',width=900, height=500)
```

A more accurate definition of **Data Science** has been extracted from a NYU paper¹ as:

- "The term “Science” implies knowledge gained by systematic study. According to one definition, it is a
   systematic enterprise that builds and organizes knowledge in the form of testable explanations and
   predictions about the universe. Data Science might therefore imply a focus around data and by
   extension, Statistics, which is a systematic study about the organization, properties, and analysis of data
   and their role in inference, including our confidence in such inference."


### Turning our attentions to Machine Learning:

It's really common over the scientific community to people present papers talking about "Machine Learning" as something esoteric or really distant from the reality. However, as we're going to see over the next sections, Machine Learning consists, basically, in Statistical models that can be implemented in almost anykind of data allowing to:
1. Predict data generation process via of algorithm learning method;
2. Describe the ocurrency of events using a probabilistic space based on a ex-ante model; 

```python
#Branchs of Data Science:
from IPython.display import Image
Image('http://www.rosebt.com/uploads/8/1/8/1/8181762/4920848_orig.png',width=500)
```

Basically, there are three branches in Machine Learning:

**1. Supervised Learning (it's provided data labels):**
- Based on ex-ante information about data structure the algorithm tries to categorize new data into the already know information;
    
**2. Unsupervised Learning (it's provided only features, not data labels):**
- The algorithm tries to predict data structure based on its intrisic information;

**3. Semi-supervised Learning (it can be provided or not data labels):**
- The algorithm forces itself to extract new information based on the already learned proccess;

### 3.1.0) Jupyter Notebook

"The Jupyter Notebook is an open-source web application that allows you to create and share documents that contain live code, equations, visualizations and narrative text. Uses include: data cleaning and transformation, numerical simulation, statistical modeling, data visualization, machine learning, and much more".

It's really interesting to use Jupyter Notebook while writing pieces of codes:

    1) You can write independent piece of codes, allowing you to manipulate as you need;
    2) It allows you to run only some parts of codes - it doesn't force you to re-run all piece of codes, whice can be a potential problem while dealing with Big Data;
    3) Jupyter Notebook provides an easy and intuitive interface for beginners combined with a good computational capability to solve most of basic Economic Analysis;
    4) It provides some graphical solutions and html possibility to share scripts;
    5) It has compability with LaTeX allowing you to write piece of formulas inside the markdown areas;

### 3.1.3) Data Analysis and Big Data Libraries

We're going to provide a brief introduction for two very important libraries in Python while dealing with Data Analysis:

#### Pandas
Pandas is one of the most used libraries for data analysis in Python. It supports a wide variety of data formats, such as txt, csv, xlsx, and one of the most used format for Economists: dta.

Most importantly, it provides tools for Big Data exploration in a more general way, but requires some careful, as it's not always efficient computationally, as it doesn't support 'paralelism' while opening and dealing with Big Data (more than 10mn of rows). Nevertheless, it has compability with SkLearn, allowing an easier approach while running small Machine Learning models (in most cases it's a better idea work with vectorized data).

#### Dask
Dask is a relatively new library, and really astonishing while loading Big Data. It supports 'paralelism', allowing a more efficient approach while loading Big Data and transforming it as our necessities. Also, you can open compressed files, resulting in less Memory Usage while working in servers.

Unfortunately, it does not load some data types such as fixed width files, very common in files provided by the Brazilian Agencies.

```python
#The most used library to read data and perform massive operations
import pandas as pd

#A very useful library to read data in paralelism - BigData contribution
import dask as dk
```
